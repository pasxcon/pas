```python
import folium

# Define two sets of four coordinates
coordinates_green = [(40.7128, -74.0060), (34.0522, -118.2437), (41.8781, -87.6298), (29.7604, -95.3698)]
coordinates_red = [(51.5074, -0.1278), (48.8566, 2.3522), (40.7128, -74.0060), (34.0522, -118.2437)]

# Create a map centered at a particular coordinate
map_center = (38.9072, -77.0369)  # Washington, D.C.
mymap = folium.Map(location=map_center, zoom_start=4)

# Draw polygons for the areas enclosed by the coordinates
folium.Polygon(locations=coordinates_green, color='green', fill=True, fill_color='green', fill_opacity=0.5).add_to(mymap)
folium.Polygon(locations=coordinates_red, color='red', fill=True, fill_color='#DAD', fill_opacity=0.5).add_to(mymap)

# Add markers for each coordinate
for coord in coordinates_green:
    folium.Marker(coord, icon=folium.Icon(color='green')).add_to(mymap)
for coord in coordinates_red:
    folium.Marker(coord, icon=folium.Icon(color='red')).add_to(mymap)

# Save the map to an HTML file
mymap.save("map_with_polygons.html")
mymap
```
